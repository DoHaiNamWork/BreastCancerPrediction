```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
```

## EDA

```python
df = pd.read_csv('breastCancer_data.csv')
df
```

Drop columns

```python
df.drop(columns={'id', 'Unnamed: 32'}, inplace=True)
```

```python
df
```

```python
df['diagnosis'].value_counts()
```

```python
df['diagnosis'] = df['diagnosis'].map({'M':1, 'B':0})
```

Splitting data into Training, Validation and Test set

```python
train_val_df, test_df = train_test_split(df, test_size=0.2, random_state=42)
train_df, val_df = train_test_split(train_val_df, test_size=0.25, random_state=42)
```

```python
target_col = 'diagnosis'
input_cols = list(train_df.columns)[1:]
```

```python
train_inputs = train_df[input_cols].copy()
train_target = train_df[target_col].copy()

val_inputs = val_df[input_cols].copy()
val_target = val_df[target_col].copy()

test_inputs = test_df[input_cols].copy()
test_target = test_df[target_col].copy()
```

```python
print(train_df.shape[0], val_df.shape[0], test_df.shape[0])
```

## Logistic Regression

```python
LR_model = LogisticRegression(solver='liblinear')

LR_model.fit(train_inputs, train_target)
```

```python
LR_model.coef_
```

```python
LR_model.intercept_
```

Accuracy

```python
print(f'Training Accuracy: {LR_model.score(train_inputs, train_target)}')
```

```python
print(f'Validation Accuracy: {LR_model.score(val_inputs, val_target)}')
```

Confusion Matrix

```python
pd.DataFrame(confusion_matrix(train_target, LR_model.predict(train_inputs)))
```

```python
pd.DataFrame(confusion_matrix(val_target, LR_model.predict(val_inputs)))
```

Hyperparameters Tuning

Function to find the best max_depth

```python
def max_depth_error(md):
    model = DecisionTreeClassifier(max_depth=md, random_state=42)
    model.fit(train_inputs, train_target)
    train_acc = model.score(train_inputs, train_target)
    val_acc = model.score(val_inputs, val_target)
    return {'Max Depth': md, 'Training Accuracy': train_acc, 'Validation Accuracy': val_acc}
```

```python
accuracy_df = pd.DataFrame([max_depth_error(md) for md in range(1, 7)])
```

```python
accuracy_df.sort_values('Validation Accuracy', ascending=False)
```

```python
DecisionTree_model = DecisionTreeClassifier(random_state=42, max_depth=4)

DecisionTree_model.fit(train_inputs, train_target)
```

```python
DecisionTree_model.score(train_inputs, train_target)
```

```python
DecisionTree_model.score(val_inputs, val_target)
```

## Comparing Models

```python
train_accuracy = [LR_model.score(train_inputs, train_target), DecisionTree_model.score(train_inputs, train_target), RF_model.score(train_inputs, train_target)]
val_accuracy = [LR_model.score(val_inputs, val_target), DecisionTree_model.score(val_inputs, val_target), RF_model.score(val_inputs, val_target)]

model_accuracy_df = pd.DataFrame({
    'Model': ['LogisticRegression', 'DecisionTree', 'RandomForest'],
    'Train_Accuracy': train_accuracy,
    'Val_Accuracy': val_accuracy
    })
```
